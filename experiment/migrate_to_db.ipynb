```python
import numpy as np
from os import listdir
import fnmatch
from datetime import datetime
from os.path import isfile, join, isdir
import pandas as pd
```

```python
def parse_filename(filename):
    chunk_vec = filename.replace(".npy", "").split("_")
    k = int(chunk_vec[2])
    priv_lev = float(chunk_vec[4])/10
    n = int(chunk_vec[8])
    method = chunk_vec[5] + chunk_vec[6]
    
    return(k, priv_lev, method, n)
```

```python
dir = "/Users/jmmoon/Documents/GitHub/LDPUts/experiment/power_ptbunif_k4"

dir_result = dir + "/result"
filelist=  [f for f in listdir(dir_result) if fnmatch.fnmatch(f, '*.npy')]

method_list = []
alpha_list = []
n_list = []
for fileitem in filelist:
    k, priv_lev, method, n = parse_filename(fileitem)
    method_list.append(method)    
    alpha_list.append(priv_lev)  
    n_list.append(n)

method_list = np.unique( np.array(method_list))
alpha_list = np.unique( np.array(alpha_list))
n_list = sorted(np.unique( np.array(n_list)))

print(method_list)
print(alpha_list)
print(n_list)

file_dict  = dict(zip(method_list, [dict(
    zip(alpha_list, [dict(
        zip(n_list, ["" for p in n_list])
    ) for alpha in alpha_list])
) for method in method_list]))


for file_now  in filelist:
    k, priv_lev, method, n = parse_filename(file_now)
    file_dict[method][priv_lev][n] = file_now
print(file_dict)
```
